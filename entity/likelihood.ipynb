```python
import pandas as pd
import spacy
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
df = pd.read_csv('result_deberta.tsv', sep='\t')
entity_list = ['CARDINAL', 'DATE', 'EVENT', 'FAC', 'GPE', 'LANGUAGE', 'LAW', 'LOC', 'MONEY', 'NORP', 'ORDINAL', 'ORG', 'PERCENT', 'PERSON', 'PRODUCT', 'QUANTITY', 'TIME', 'WORK_OF_ART']
for entity_kind in entity_list:
    df[entity_kind] = 0
    df[entity_kind + '_text'] = ''
print(df[4:])
```

```python
def find_insert_text(str1, str2):
    str1_list = str1.split(' ')
    str2_list = str2.split(' ')
    i = 0
    j = 0
    res = ''
    for j in range(len(str2_list)):
        if str1_list[i] != str2_list[j]:
            res += str2_list[j] + ' '
        else:
            i += 1
        if i == len(str1_list):
            break
    return res
```

```python
for index, row in df.iterrows():
    text_a = row['text_a']
    text_b = row['text_b']
    insert_text = find_insert_text(text_a, text_b)
    doc = nlp(insert_text)
    if len(doc.ents) > 0:
        for ent in doc.ents:
            row[ent.label_] += 1
            if row[ent.label_] == 1:
                row[ent.label_ + '_text'] += ent.text
            else:
                row[ent.label_ + '_text'] += ',' + ent.text
df.to_csv('result_deberta_entity_likelihood.tsv', sep='\t', index=False)
```
